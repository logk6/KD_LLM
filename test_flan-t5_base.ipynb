```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, GenerationConfig

# CHECKPOINT = "D:\DUNG\MasterProgram\IT5412_LLM\Project\TinyLLM_source\TinyLLM-OBQA"

model_name = 'google/flan-t5-base'
flan_t5_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)


generation_config = GenerationConfig(max_new_tokens=64, do_sample=True, temperature=0.7)
```

```python

question = """Question: A person who forgets to drink water on a hot day may feel dizzy because dehydration reduces what in the body?
Options:
A. Hair growth rate
B. Blood volume
C. Bone strength
D. Vision clarity permanently
Answer with the correct choice
Ans: B
"""

inputs = tokenizer(question, return_tensors="pt")
output_seq = flan_t5_model.generate(inputs["input_ids"], generation_config=generation_config)
output = tokenizer.decode(output_seq[0], skip_special_tokens=True)
print(output)
```
